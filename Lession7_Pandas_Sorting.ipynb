```python

#Important Methods for Sorting DataFrames
#sort_values() - Sorts the DataFrame by the values of one or more columns.
#sort_index() - Sorts the DataFrame by its index (row labels).
#nlargest(n) - Returns the n largest values in a column.
#nsmallest(n) - Returns the n smallest values in a column.

import pandas as pd
inpFilePath ='/Users/sam/git/Pandas_Data/Input/Data_2019/'
df = pd.read_csv(inpFilePath+'survey_results_public.csv')
schema_df = pd.read_csv(inpFilePath+'survey_results_schema.csv')
pd.set_option('display.max_columns', 85) #to display all columns in the dataframe
pd.set_option('display.max_rows', 85)   #to ee all rows in the dataframe of schema
pd.set_option('display.max_colwidth', 100) #to see the full description of the columns in the schema dataframe

people = {
    "first": ["Corey", "Jane", "John","Adam"],
    "last": ["Schafer", "Doe", "Doe","Doe"],
    "email": ["Corey.Schafer@gmail.com", "Jane.Doe@gmail.com", "John.Smith@gmail.com","Adam.Doe@gmail.com"]
}
df1 = pd.DataFrame(people)
df1
```

```python
df1.sort_values(by='last') #if there is a tie in the last name, then it will sort by whosoever comes first in the dataframe. 
#In this case, Jane will come before John and Adam because Jane is before them in the dataframe.
```

```python
df1.sort_values(by='last', ascending=False)
```

```python
df1.sort_values(by=['last','first'],ascending=[False,True],inplace=True)
df1
```

```python
df1.sort_index() #will sort based on row labels , here its index
#but this will not update the dataframe because we have not set inplace to True. 
# So, we will have to set inplace to True to update the dataframe.S
```

```python
df1['last']
```

```python
#we can also sort the series i.e. columns of the dataframe using sort_values() method.
df1['last'].sort_values() #will sort in ascending order by default.
```

```python
df['ConvertedComp'].nlargest(10) #for all the values in the column ConvertedComp, it will return the 10 largest values. of convertedComp.
```
