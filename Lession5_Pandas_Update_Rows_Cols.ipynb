```python
def make_uppercase(string):
    return string.upper()

df1['email'] = [make_uppercase(x) for x in df1['email']] 
df1['email']   
```

```python
def update_email(email):
    return email.lower()

#df1['email'].apply(update_email) 
# this will return a new series
#  but will not update the original dataframe. 
# We need to assign it back to the original dataframe
#  to update it.
df1['email'] = df1['email'].apply(update_email)
df1['email']
```

```python
#Lamda Functions
df1['email'] = df1['email'].apply(lambda x: x.upper())
df1['email']

#Lambda functions (or anonymous functions) are best 
# used for short, single-use, and simple operations 
# where defining a named function (def) would be 
# overkill. They are ideal as arguments to higher-order
#  functions—like map(), filter(), or sorted()—or 
# for creating small, disposable logic to be used once.


#Filtering a list in Python: 
#filtered = list(filter(lambda x: x % 2 == 0, numbers))
#Sorting with a key in Python: 
#data.sort(key=lambda x: x.name)
#Simple operations in C++: 
#auto add = [](int a, int b) { return a + b; }; 
```

```python
 #1. df1.applymap(lambda x: x.lower())

#2. df1.applymap(str.lower)

#above 2 are same this str.lower will also work 
# because it is a string method and 
# applymap will apply it to each element 
# in the dataframe.
#however it will not work if there 
# are any non-string values in the dataframe.
#in that case we can use a lambda function to check 
#if the value is a string before applying the lower method.
#something like below
df1.applymap(lambda x: x.lower() if isinstance(x, str) else x)

```

```python
#map is used to substitute each value in a Series with another value,
# that may be derived from a function, a dictionary, or a Series.

#df1['first'] is a series object and 
# we can use map method on it to 
# replace the values in the series 
# with the values in the dictionary.

df1['first'].map({'Corey': 'Chris', 'Jane': 'Mary'})
#This will replace 'Corey' with 'Chris' 
# and 'Jane' with 'Mary in the 'first' 
# column of the dataframe pd1. 
#The values that are not in the dictionary 
# will be replaced with NaN.
```

```python
#If we want to keep the values that are not in the dictionary
# then replace method is used instead of map method.
df1['first'].replace({'Corey': 'Chris', 'Jane': 'Mary'})
#This will replace 'Corey' with 'Chris'
# and 'Jane' with 'Mary in the 'first'
#df1['first'] = df1['first'].replace({'Corey': 'Chris', 'Jane': 'Mary'})
# this will update the values in the dataframe because 
# we are assigning the result back to the original dataframe.
```
